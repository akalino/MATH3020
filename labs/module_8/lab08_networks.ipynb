When we looked at paths and traversals, it did not really matter which node served as the start or the end; as long as the path length was the same, we could use it to compute some *invariants* about the graph, such as the radius, diameter, and center.

Computing invariants (simply, metrics that don't change under deformations (for example, an isomorphism preserves all invariants)) is a fundamental notion in mathematics. They help us classify, organize, and generalize mathematical objects, allowing for links between seemingly disparate fields of study. 

The idea we will build toward is to compare these metrics of real-world graphs to different families of graphs, with the goal of using statistical testing to show (or reject null hypotheses) that the real-world graph structures differ from pure randomness.

## Part 2: Direction changes the story

Now we move from undirected graphs to directed graphs.

A directed edge from \(u\) to \(v\) is not the same as an edge from \(v\) to \(u\).
This matters when the graph models things like traffic, communication, influence, or hyperlinks.

```python
D = sample_directed_graph()
digraph_summary(D)
```

Here is where we introduce the idea of in-degree and out-degree. The in-degree is the number of edges (edges coming in, sometimes in network science circles, called links; think of social platforms) and out-degrees (edges going out).

```python
draw_digraph(D, "Sample directed graph")
```

```python
print("Original graph \n")
print(adjacency_matrix_table(G))
print("Directed graph \n")
print(adjacency_matrix_table(D))
```

## Part 4: Weighted graphs

We can continue with directed graphs by appending additional information to the edges. For the class of weighted graphs, we can add a numerical value to each edge (as a weight) that may represent:

- distance
- cost
- travel time
- strength of interaction
- frequency of communication

The same graph structure can tell a very different story once weights are included.

This is not the only extension we can apply to edges. We can include additional information, including edge types (friends, enemies, co-workers). We can similarly imbue the vertices with *properties*, such as a person's age, hobbies, location; graphs of this type are commonly called *knowledge graphs*, and have important applications in social networks, database integration, and identifying fraud networks (see https://medium.com/neo4j/graph-visualization-of-panama-papers-data-in-neo4j-9c08ca17039c)

We can add weights to the edges of a NetworkX graph by supplying the edge as a tuple before, but adding a third element for the numeric weight. Take a look at the sample_weighted_graph function in the helpers for this lab.

```python
W = sample_weighted_graph()
weighted_graph_summary(W)
```

```python
draw_weighted_graph(W, "Sample weighted graph")
```

```python
weighted_edge_table(W)
```

## Part 5: Not all shortest paths mean the same thing

In an unweighted graph, a shortest path means the fewest number of edges.

In a weighted graph, a shortest path usually means the path of minimum total weight.

```python
shortest_path_weighted(W, "A", "E")
```
