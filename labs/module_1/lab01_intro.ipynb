## Build the graph

We will take a set of vertices and label them using a Python string.
Let's say we have five people: Alice, Bob, Carol, Dana, and Eddie, who work with one another.
Alice is the boss of Eddie and Dana, Carol is the boss of Alice, and Bob reports to both Eddie and Dana.
We can represent the "works for" relationship by making a Python list (square brackets) of pairs of names (called a tuple in Python).
We will store this edge list in a variable called "works_for":

```python
works_for = [("Eddie", "Alice"), ("Dana", "Alice"), ("Alice", "Carol"), 
             ("Bob", "Eddie"), ("Bob", "Dana")]
```

The edge list above is just a list of pairs. Now turn it into a graph object.

```python
G = build_graph_from_edge_list(works_for)
```

We can then print a summary of the graph object.

```python
summary = basic_summary(G)
print(summary)
```

The summary function provided returns a Python dictionary. 
We will see how to access information from the dictionary later in the course.
Thus far, we have seen four of the main data types in Python: strings, tuples, lists, and dictionaries.

## Degree information

A first useful question is: how many connections does each vertex have?


```python
degree_table(G)
```

From the degree table, we can see that Alice has three connections to her co-workers, while Carol has only one connection.
We can then view this information as a histogram that describes the count of people having each vertex degree.
The output will again be a dictionary: the first entry (called the key) represents the degree, the second entry (called the value) represents the count for that particular degree.

```python
degree_histogram(G)
```

## Draw the graph

This gives a quick visualization in Python.


```python
draw_graph(G)
```
