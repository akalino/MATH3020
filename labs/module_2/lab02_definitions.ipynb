### Paths
Since the graph is now connected, there is a path between every vertex.
Let's check that this is true. The code block below will iterate through 
all the pairs of vertices in the graph and compute the path length between them.

This is an example of a Python function. It takes an arguement (_g) that allows us 
to pass in a graph object.
The function will then loop over all the vertices (twice for all the pairs)
and compute the path length between all the pairs.

```python
def find_path_lengths(_g):
    """ A function to find the path length between all pairs"""
    vertices = _g.nodes()
    for v in vertices:
        start_node = v
        remaining = vertices - v
        for j in remaining:
            end_node = j
            p_len = len(nx.shortest_path(_g, start_node, end_node))
            print(f"Path length between {start_node} and {end_node} is {p_len}")
```

```python
find_path_lengths(G2)
```

We can use this function to define another function.
This second function will find the longest path of all the shortest paths computed above.
This metric is called the diameter of the graph, and we will see this again in the future.

```python
def get_diameter(_g):
    shortest_path_lengths = []
    vertices = _g.nodes()
    for v in vertices:
        start_node = v
        remaining = vertices - v
        for j in remaining:
            end_node = j
            p_len = len(nx.shortest_path(_g, start_node, end_node))
            shortest_path_lengths.append(p_len)
        return max(shortest_path_lengths)
```

```python
get_diameter(G2)
```

We can additionally add this to our dictionary of metrics by adding a new key/value pair.

```python
gm2 = graph_summary(G2)
gm2['diameter'] = get_diameter(G2)
print(gm2)
```
