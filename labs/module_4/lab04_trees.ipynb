### Number of vertices and edges

Let's continue with the same example. Note the number of vertices and edges in a tree. We can learn a lot about this relationship experimentally to form a conjecture about the relationship *prior* to searching for a proof. Let's first look at this relationship for the same tree $T$ we have been using.

```python
n = T.number_of_nodes()
m = T.number_of_edges()

print("number of vertices =", n)
print("number of edges    =", m)
print("m = n - 1 ?", m == n - 1)
```

Obviously, a single example does not guaruntee this will hold for every tree in all edge cases. But, we can collect some more supporting evidence the relationship holds. At the same time, there will always be *edge cases* that "buck the pattern"; this is where mathematical proof, rooted in logic, become the most valuable tool.

To generate this supporting evidence, we can encode the definition of a tree into a function that depends solely on the number of vertices. Let's do this in a loop and look at the results.

```python
n_vert = [3, 4, 5, 10, 20, 50, 100, 200, 500]
n_edge = []
for n in n_vert:
    nt = generate_random_tree(n)
    n_edge.append(nt.number_of_edges())
out = pd.DataFrame({'n_vert': n_vert,
                    'n_edge': n_edge})
print(out)
```

It's looking like the pattern still holds! Even at a "small" number of $n$ vertices, the number of edges stays at $n-1$. This is shaping up to be an opportunity for a proof. 

Let's consider one other opportunity for a proof: are all trees of size $n$ isomorphic to one another?
