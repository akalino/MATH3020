## Part 3: A planar graph can still be drawn badly

A graph may be planar even if one particular drawing has many crossings.

Try drawing the same graph in multiple ways. The NetworkX package has several built-in layouts that are commonly used. We'll look at two: the spring layout and the circular layout. Check out the documentation at https://networkx.org/documentation/stable/reference/drawing.html for a larger discussion on the other available layouts. However, NetworkX even states that it's not geared toward graph visualization, especially at larger scales. This is where Gephi will later shine.

```python
G_planar = wheel_graph(6)
graph_summary(G_planar)
```

```python
G_planar = wheel_graph(6)
graph_summary(G_planar)
```

```python
draw_graph_spring(G_planar, "Wheel graph with spring layout")
draw_graph_circular(G_planar, "Wheel graph with circular layout")
```

Based on the above, we can't always rely on a drawing of the graph to determine planarity. Poor drawings may exhibit crossings, but there may be a way of 'untangling' those crossings to arrive at a planar graph. This is where Kuratowski's theorem really comes into play: if we absolutely know that $K_5$ or $K_{3,3}$ is a subgraph of the entire graph $G$, there is no way to ever 'untangle' $G$ to be planar.

## Part 4: Edge bounds

For a simple planar graph with $v \ge 3$, one important necessary condition is

$$
e \le 3v - 6.
$$

This condition does not prove planarity, but it can rule out planarity.

Use it on a few examples.

```python
print(f"Cycle graph stats: {edge_bound_report(G_cycle)}")
print(f"K_4 graph stats: {edge_bound_report(G_k4)}")
print(f"K_5 graph stats: {edge_bound_report(G_k5)}")
print(f"K_3,3 graph stats: {edge_bound_report(G_k33)}")
```

The bounds formula is incredibly useful in taking simplistic graph statistics and utilizing them to reject planarity for a large number of graphs. This narrows the search space to find a planar embedding of a particular graph.
